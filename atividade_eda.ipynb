Histograma

```python
df['home_score'].value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Quantidade de gols da casa por Partida")
plt.ylabel('Quantidade de partidas')
plt.xlabel('Quantidade de gols');
```

Podemos constatar que houve poucas goleadas (3 gols ou mais), sendo assim, foi um campeonato com partidas mais equilibradas

```python
from collections import defaultdict
vitorias = defaultdict(int)
empates = defaultdict(int)
derrotas = defaultdict(int)

for _, row in df.iterrows():
    home = row['home_team']
    away = row['away_team']
    home_score = row['home_score']
    away_score = row['away_score']

    if home_score > away_score:
        vitorias[home] += 1
        derrotas[away] += 1
    elif home_score < away_score:
        derrotas[home] += 1
        vitorias[away] += 1
    else:
        empates[home] += 1
        empates[away] += 1
```

```python
teams = sorted(set(vitorias.keys()).union(empates.keys()).union(derrotas.keys()))
data_summary = pd.DataFrame({
    'team': teams,
    'wins': [vitorias[t] for t in teams],
    'draws': [empates[t] for t in teams],
    'defeats':[derrotas[t] for t in teams]
})

data_summary = data_summary.sort_values(by='wins', ascending=False).reset_index(drop=True)
data_summary.head()
```

```python
data_long = pd.melt(data_summary.head(), id_vars='team', value_vars=['wins', 'draws', 'defeats'],var_name='resultado', value_name='quantidade')

plt.figure(figsize=(14, 6))
sns.barplot(data=data_long, x='team', y='quantidade', hue='resultado', palette='Set2')
plt.title("Vitórias e Empates por Seleção")
plt.xlabel("Seleção")
plt.ylabel("Quantidade")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Percebe-se que a Espanha, por ser a vencedora do campenato, obteve disparadamente, o melhor desempenho, não perdendo nenhum jogo, percebemos que os times do top 4, obteve um desempenho de vitórias superior ao de derrotas e empates.
